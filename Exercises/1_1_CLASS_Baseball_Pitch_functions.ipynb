```python
def convert_para(value,old,quality):
    foot_to_m = 0.3048
    mile_to_m = 1609.34
    hour_to_s = 3600
    if quality == 'velocity':
        if old == 'british':
            conv = (value * mile_to_m)/hour_to_s
        else:
            conv = (value * hour_to_s)/mile_to_m
    if quality == 'distance':
        if old == 'british':
            conv = value*foot_to_m
        else:
            conv = value/foot_to_m
    return conv

def compute_time(dist, v_i):
    return dist/v_i

def compute_target(t,rel_h,cat_h,a):
    drop = (1/2)*a*(t**2)
    ideal_drop = rel_h - cat_h
    target = ideal_drop - drop
    return target
```

```python
# Define parameters
dist = 60.6     # distance to the plate, feet
rel_h = 8       # release height, including height of the mound, feet
cat_h = 2.5     # height of the catcher's mitt, feet
v_i = 70        # initial velocity of pitch, mph
a = 9.81        # gravitational acceleration, m/s^2

dist = convert_para(dist,'british','distance')
rel_h = convert_para(rel_h,'british','distance')
cat_h = convert_para(cat_h,'british','distance')
v_i_m = convert_para(v_i, 'british','velocity')
t = compute_time(dist,v_i_m)
target = compute_target(t,rel_h,cat_h,a)
target_b = convert_para(target,'metric','distance')

print("When pitched horizontally at a velocity of", v_i,  " mph, the pitch reaches homeplate ",
      target_b, " feet 'above' the catcher's mitt.")


```
