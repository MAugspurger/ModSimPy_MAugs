```python
import pandas as pd; import numpy as np

def run_simulation(system, change_func):
    t_array = np.arange(0, system['t_end']+1, system['dt'])
    results = pd.Series(index=t_array,dtype=object)
    results[0] = system['T_init']
    for t in t_array:
        T = results[t]
        delta = change_func(t, T, system)
        T = T + delta
        results[t+1] = T
    system['T_final'] = results.iloc[-1]
    return results

def change_func(t, T, system):
    r, T_env, dt = system['r'], system['T_env'], system['dt']
    delta = -r * (T - T_env) * dt
    return delta
```

```python
T_init = 90.0;  volume = 300; r = 0.05; t_end = 30; T_env = 22; T_goal = 70.0
coffee_system = dict(T_init=T_init, T_env=T_env,volume=volume, r=r, t_end=t_end, dt=1)
results = run_simulation(coffee_system,change_func)
results.plot(xlabel='time (s)', ylabel='Temperature');
```
