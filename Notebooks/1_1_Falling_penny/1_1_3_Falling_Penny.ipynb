## Modeling the Velocity of a Penny Falling

Now we're going to create a model to determine the final speed and time required for the penny to fall.

<br>

The statistician George Box famously said "All models are wrong, but
some are useful." The puzzle that modelers need to solve is how to create a model that combines accuracy (usefulness) and inaccuracy (efficiency) in the right mix: we want a model that gives us a usable answer without wasting time and resources.

### Investigation and Abstraction

In your investigation in the exercises of the last notebook (i.e. observing the *Mythbusters* video), you probably decided that drag was an important factor for the velocity of the penny, and you were correct.   But sometimes we implement a model that we know is too simple just to make sure our implementation is successful, and then we iterate on that model to add more complexity.

<br>

We'll do that here.  We're going to reduce our list of possible factors affecting the final velocity of the penny to two factors: gravitational force and height.  We'll assume the effect of all other factors is *negligible* (that is, they have only small effects), and so we will ignore them.  Engineers do this all the time, no matter how much the mathematicians resent it üòè

<br>

So we have completed the first half of abstraction: the selection of our key factors.  The second part of abstraction is defining the *relationships* between the key factors.  In the last notebook, we dropped a penny from a ladder (in our heads, anyway) and hypothesized that the relationship between time and velocity of a falling penny was linear:

<br>

$$v=Ct$$

<br>

Another form of "investigation" would be to look at the work of previous scientists.  A quick look at an intro physics book would reveal this work, in the form of the *kinematic equations*.  These equations tell us about the relationships between time, height and velocity for an object in free-fall (when drag is negligible).  

<br>

Happily, these equations tell us that our ladder investigation was valid!  The kinematic equation relating time and velocity in free-fall is:

$$v = g t$$

This model tells us that if the initial velocity is 0, the velocity after time $t$ is a function of the gravitational constant $g$, which at sea level is approximately $g=9.81$ $m/s^2$.

<br>

A second kinematic equation relates time and distance fallen $x$:

$$x = g t^2 / 2$$

<br>

With a little math magic, we can combine these two equations to find the relationship between our two factors and the final velocity of the penny:

<br>

$$ v = \sqrt{2gx}$$

<br>

This equation is our abstracted model for the final velocity of a penny-dropped from a height $x$ when drag is negligible.

### Implementation and Validation

"Implementation" in this case is pretty straight-forward: we can just take the values from our particular situation and plug them into our model equation.

<br>

We could do this with a calculator, but let's use Python. First we'll create a variable to represent acceleration due to gravity in meters per second squared (m/s$^2$).

```python
g = 9.8
```

A *variable* is a name that corresponds to a value. In this example, the name is `g` and the value is the number `9.8`.  When you run the above cell (click the `play` button or press `shift-enter`), nothing seems to happen.  But in fact something does: the code *assigns* the value `9.8` to the variable `g`.

<br>

Suppose we let the penny drop for $3.4$ seconds (s).  I'll create a variable to represent this time:

```python
t = 3.4
t
```

Notice that when I add the second line `t`, running the cell produces an output.  This doesn't change the variable at all: it just tells the computer to display the value of `t`.

<br>

Now we can compute the velocity of the penny after `t` seconds, and assign the answer to a variable named `v`:

```python
v = g * t
```

Python uses the symbol `*` for multiplication.  The other arithmetic operators are `+` for addition, `-` for subtraction, `/` for division, and `**` for exponentiation (like $5^2$).  The variable `v` now has been assigned a value.  We can display its value like this:

```python
v
```

After $3.4$ s, the velocity of the penny is about $33$ m/s (ignoring air resistance).  Now let's see how far it would travel during that time.  Remember from above that our equation for distance travelled is $x = g t^2 / 2$:

```python
x = g * ((t**2)/ 2)
x
```

It would travel about $56$ m.  Now, going in the other direction, let's compute the time it takes to fall 381¬†m, the height of the Empire State Building.

```python
h = 381
```

For this computation, we need the square root function, `sqrt`, which is provided by a *library* called NumPy, which contains functions related to the manipulation of numbers.  NumPy is built into Colab.  But before we can use it in our notebook,we have to *import* it.  Importing a library or function means we make it available to use, even though the code for that function is not in this particular notebook:

```python
from numpy import sqrt
```

Again using the kinematic equation that we used above, we can write:

```python
t = sqrt(2 * h / g)
t
```

With no air resistance, it would take about $8.8$ s for the penny to reach the sidewalk.

```python
v = g * t
v
```

And its velocity on impact would be about $86$ m/s, or about 190 miles per hour.  That sounds like it could hurt.  We don't need to do much here in terms of validation, though.  Our intuition (aided by the *Mythbusters*) tells us that a penny travelling at 190 mph would quickly be slowed down by drag.  

<br>

This model needs some iteration....

### Iteration

We are now ready to enter the implementation/ validation/ iteration feedback loop!  

<br>

As the *Mythbusters* guys showed, ignoring air resistance leads to major errors. So we want to iterate our model: that is, we want to add in some new complication that will make the model more troublesome, but happily more accurate.  It turns out that once
the penny gets to about 29¬†m/s, the upward force of air resistance
equals the downward force of gravity, so the penny stops accelerating.
This is the *terminal velocity* of the penny in air.

<br>

Once the penny reaches terminal velocity, it doesn't matter how much farther it falls; it hits the sidewalk at about 29¬†m/s (65 mph).
That's much less than 86¬†m/s, as the simple model predicts.
