The following function is a possible answer to the final exercise in Notebook 1,.2.3. which asks you to record the first time that an unhappy customer tries to find a bike but the bike rack is empty.  `t_first_empty` is added to `state` for this change function; the time of the first unhappy customer is stored in this variable.  Let's document this function:

```python
def change_func(state, p1, p2):
    state.clock += 1
    
    if flip(p1):
        if state.moline == 0:
            state.moline_empty += 1
        else:
            state.moline -= 1
            state.augie += 1
    
    if flip(p2):
        if state.augie == 0:
            state.augie_empty += 1
        else:
            state.augie -= 1
            state.moline += 1
        

    if state.t_first_empty < 0:
        if state.augie_empty + state.moline_empty > 0:
            state.t_first_empty = state.clock
```
