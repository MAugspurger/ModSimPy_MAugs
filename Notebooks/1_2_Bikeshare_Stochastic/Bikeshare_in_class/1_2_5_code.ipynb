```python
def change_func(state, ptm, pta):
    if npr.random() < ptm:
        print("Bike to Moline")
        state.moline += 1
        state.augie -= 1
    if npr.random() < pta:
        print("Bike to Augie")
        state.moline -= 1
        state.augie += 1
    return state

def run_simulation(ptm, pta, iAug, iMol, num_steps):
    state = pd.Series(dict(augie=iAug,moline=iMol),name="Number of Bikes")
    state.index.name = 'Location'
    for i in range(num_steps):
        state = change_func(state, ptm, pta)
    return state
```

```python
# Runtime code
import pandas as pd
import numpy.random as npr
bikeshare = run_simulation(0.5, 0.4, 10, 2, 15)
print("There are", bikeshare.augie, "bikes at Augie")
```
