```python
def change_func(ptm, pta):
    if npr.random() < ptm:
        print("Bike to Moline")
        bikeshare.moline += 1
        bikeshare.augie -= 1
    if npr.random() < pta:
        print("Bike to Augie")
        bikeshare.moline -= 1
        bikeshare.augie += 1
```

```python
# Runtime code
import pandas as pd
import numpy.random as npr
state = dict(augie=10,moline=2)
bikeshare = pd.Series(state,name='Number of Bikes')
bikeshare.index.name = 'Location'
change_func(0.5,0.4)
print("There are", bikeshare[0], "bikes at Augie")
```
