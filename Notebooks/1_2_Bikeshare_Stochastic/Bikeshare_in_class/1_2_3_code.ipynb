```python
def change_func(state, ptm, pta):
    if npr.random() < ptm:
        print("Bike to Moline")
        state.moline += 1
        state.augie -= 1
    if npr.random() < pta:
        print("Bike to Augie")
        state.moline -= 1
        state.augie += 1
    return bikeshare
```

```python
# Runtime code
import pandas as pd
import numpy.random as npr
init_state = dict(augie=10,moline=2)
bikeshare = pd.Series(init_state,name='Number of Bikes')
bikeshare.index.name = 'Location'
bikeshare = change_func(bikeshare, 0.5,0.4)
print("There are", bikeshare[0], "bikes at Augie")
```
